## Task 4

Convert the RGB image into the HSV color space. Rotate the colors in the HSV color space, convert it back to the RGB color space and plot it together with its histogram. For this task, you are only allowed to use NumPy.

More information about the HSV color space: https://en.wikipedia.org/wiki/HSL_and_HSV

````python
# Chatgpt code:
def pixel_rgb2hsv(rgb):
    # Normalize the RGB values to [0, 1]
    rgb = np.array(rgb) / 255.0
    r, g, b = rgb[0], rgb[1], rgb[2]
    
    # Find the minimum and maximum values among R, G, B
    cmax = np.max(rgb)
    cmin = np.min(rgb)
    delta = cmax - cmin
    
    # Hue calculation
    if delta == 0:
        h = 0
    elif cmax == r:
        h = (60 * ((g - b) / delta) + 360) % 360
    elif cmax == g:
        h = (60 * ((b - r) / delta) + 120) % 360
    elif cmax == b:
        h = (60 * ((r - g) / delta) + 240) % 360
    
    # Saturation calculation
    if cmax == 0:
        s = 0
    else:
        s = delta / cmax
    
    # Value calculation
    v = cmax
    
    return (h, s, v)

def rgb2hsv(image):
    # Convert the entire image (RGB) to HSV
    hsv_image = np.zeros_like(image, dtype=np.float32)
    
    for i in range(image.shape[0]):  # iterate over each row (height)
        for j in range(image.shape[1]):  # iterate over each column (width)
            rgb_pixel = image[i, j]
            hsv_pixel = pixel_rgb2hsv(rgb_pixel)
            hsv_image[i, j] = hsv_pixel
            
    return hsv_image



def depr_hsv2rgb(image):
    '''
    Convert the hsv image into rgb color space and return it.

    image -- rgb image
    '''
    # TODO: implement function

    rgb = image

    for x in range(image[:,:,0].shape[0]):
        for y in range(image[:,:,0].shape[1]):

            h = image[:,:,0][x][y]
            s = image[:,:,1][x][y]
            v = image[:,:,2][x][y]

            if s == 0.0:
                rgb[:,:,0][x][y], rgb[:,:,1][x][y], rgb[:,:,2][x][y] =  v, v, v
                return rgb.astype(np.uint8)
            i = int(h*6.0) # XXX assume int() truncates!
            f = (h*6.0) - i
            p = v*(1.0 - s)
            q = v*(1.0 - s*f)
            t = v*(1.0 - s*(1.0-f))
            i = i % 6
            if i == 0:
                rgb[x,y,0], rgb[x,y,1], rgb[x,y,2] =  v, t, p
                return rgb.astype(np.uint8)
            if i == 1:
                rgb[x,y,0], rgb[x,y,1], rgb[x,y,2] = q, v, p
                return rgb.astype(np.uint8)
            if i == 2:
                rgb[x,y,0], rgb[x,y,1], rgb[x,y,2] = p, v, t
                return rgb.astype(np.uint8)
            if i == 3:
                rgb[x,y,0], rgb[x,y,1], rgb[x,y,2] = p, q, v
                return rgb.astype(np.uint8)
            if i == 4:
                rgb[x,y,0], rgb[x,y,1], rgb[x,y,2] = t, p, v
                return rgb.astype(np.uint8)
            if i == 5:
                rgb[x,y,0], rgb[x,y,1], rgb[x,y,2] = v, p, q
                return rgb.astype(np.uint8)


def pixel_hsv2rgb(hsv):
    h, s, v = hsv
    
    # If saturation is 0, then it's a shade of gray
    if s == 0:
        return np.array([v, v, v])  # R = G = B = V
    
    # Normalize hue to be in the range [0, 360]
    h = h % 360
    
    # Chroma computation
    c = v * s
    x = c * (1 - abs(((h / 60) % 2) - 1))
    m = v - c
    
    # Determine RGB components based on Hue
    if 0 <= h < 60:
        rgb = np.array([c, x, 0])
    elif 60 <= h < 120:
        rgb = np.array([x, c, 0])
    elif 120 <= h < 180:
        rgb = np.array([0, c, x])
    elif 180 <= h < 240:
        rgb = np.array([0, x, c])
    elif 240 <= h < 300:
        rgb = np.array([x, 0, c])
    else:
        rgb = np.array([c, 0, x])
    
    # Convert to RGB by adding m
    rgb = (rgb + m) * 255
    return np.clip(rgb, 0, 255).astype(np.uint8)

def hsv2rgb(hsv_image):
    # Convert the entire HSV image back to RGB
    rgb_image = np.zeros_like(hsv_image, dtype=np.uint8)
    
    for i in range(hsv_image.shape[0]):  # iterate over each row (height)
        for j in range(hsv_image.shape[1]):  # iterate over each column (width)
            hsv_pixel = hsv_image[i, j]
            rgb_pixel = pixel_hsv2rgb(hsv_pixel)
            rgb_image[i, j] = rgb_pixel
            
    return rgb_image

plt.imshow(hsv2rgb(rgb2hsv(image)))
````
